# 1. Restaurant Dataset 

```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import seaborn as sns
import re
sns.set_style('darkgrid')
```

```python
#reading the dataset
df=pd.read_csv("zomato.csv")
df.head() # prints the first N rows of a DataFrame
```

```python
restaurant = df.copy()
```

```python
restaurant.head()
```

```python
restaurant = restaurant.drop(['url','dish_liked','reviews_list','location','votes'],axis=1)
```

```python
restaurant.head()
```

```python
restaurant = restaurant.groupby(['name','address','phone','online_order','rest_type','cuisines','listed_in(type)','book_table']).first().reset_index()
```

```python
restaurant
```

```python
restaurant.to_csv('restaurant.csv')
```
